# slic the dataset

```python
data=dataset.drop(['BodyFat','Weight','Height','Neck','Chest','Abdomen','Hip','Thigh','Knee','Ankle','Biceps','Forearm','Wrist'],axis=1)
```

# let reshape our dataset to a dataframe

```python
x=data.iloc[:,0].values.reshape(-1,1)
y=data.iloc[:,-1].values.reshape(-1,1)
```

# import libaray training set and test

```python
from sklearn.model_selection import train_test_split
```

# splitting the dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                              test_size=0.1,
                                              random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

# apply linear regression

```python
lm=LinearRegression()
```

# filter the values of regression

```python
lm.fit(x_train,y_train)
```

```python
y_predict=lm.predict(x_test)
```

```python
y_predict
```
